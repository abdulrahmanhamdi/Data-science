```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python


# Load the image
Irgb = Image.open('ai.jpeg')  # Change 'sample_image.jpeg' to the image you want to process
Irgb = Irgb.convert('RGB')  # Ensure the image is in RGB format
plt.figure()
plt.imshow(Irgb)
plt.title("Original Image")
plt.show()

# Convert the RGB image to grayscale
Igray = Irgb.convert('L')  # Convert to grayscale
Igray = np.array(Igray)
plt.figure()
plt.imshow(Igray, cmap='gray')
plt.title("Grayscale Image")
plt.show()

# Get the size of the image
M, N = Igray.shape

# Initialize the histogram
histogram = np.zeros(256)

# Calculate the histogram
for r in range(M):
    for c in range(N):
        pixel_value = Igray[r, c]
        histogram[pixel_value] += 1

# Display the histogram
plt.figure()
plt.stem(histogram)  # Removed 'use_line_collection' argument
plt.title("Histogram")
plt.show()

# Apply a threshold to segment the image
threshold = 128
Ithresholded = np.where(Igray >= threshold, 255, 0).astype(np.uint8)

# Show the thresholded image
plt.figure()
plt.imshow(Ithresholded, cmap='gray')
plt.title(f"Thresholded Image (Threshold = {threshold})")
plt.show()
```
