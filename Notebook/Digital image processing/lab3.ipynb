```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Read the image
Irgb = cv2.imread('ai.jpeg')
Irgb = cv2.cvtColor(Irgb, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for displaying

# Display the RGB image
plt.figure()
plt.imshow(Irgb)
plt.title('Original Image')
plt.show()

# Convert to grayscale and calculate the standard deviation
Igray = cv2.cvtColor(Irgb, cv2.COLOR_RGB2GRAY).astype(np.float64)
std_1 = np.std(Igray)

# Display grayscale image
plt.figure()
plt.imshow(Igray, cmap='gray')
plt.title('Grayscale Image')
plt.show()

# Modify the grayscale image and calculate the new standard deviation
Igray = 1.3 * Igray
std_2 = np.std(Igray)

# Display the modified grayscale image
plt.figure()
plt.imshow(Igray, cmap='gray')
plt.title('Modified Grayscale Image')
plt.show()

# Display histogram of the grayscale image
plt.figure()
plt.hist(Igray.ravel(), bins=256, range=[0,256])
plt.title('Histogram of Grayscale Image')
plt.show()

# Thresholding
ort1 = np.mean(Igray)
Igray[Igray < 125] = 0
Igray[Igray >= 125] = 1

# Display the binary image
plt.figure()
plt.imshow(Igray, cmap='gray')
plt.title('Binary Image')
plt.show()

# Calculate new standard deviation and display the image
std_1 = np.std(Igray)
Igray = Igray + 40  # Adding constant to the binary image

ort2 = np.mean(Igray)
std_2 = np.std(Igray)

# Display the final modified image
plt.figure()
plt.imshow(Igray, cmap='gray')
plt.title('Final Modified Image')
plt.show()
```
